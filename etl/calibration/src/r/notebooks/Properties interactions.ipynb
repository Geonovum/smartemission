# s.co2 <> s.o3resistance -> O3_Waarden

```r
ggplot(df %>% sample_n(10000) %>% filter(!is.na(O3_Waarden)), aes(s.co2, s.o3resistance, color=O3_Waarden)) + 
geom_point(size = .5, alpha = .5) +
xlim(750, 1300) +
scale_color_gradientn()
```
